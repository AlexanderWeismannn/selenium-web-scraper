```python
import pandas as pd
from selenium import webdriver #launch browser
from selenium.webdriver.common.by import By #search
from selenium.webdriver.support.ui import WebDriverWait #wait for page load
from selenium.webdriver.support import expected_conditions as EC #check for page load
from selenium.common.exceptions import TimeoutException #handle timeouts

def create_webdriver():
    return webdriver.Chrome()

browser = create_webdriver()
browser.get("https://github.com/collections/machine-learning")
```

```python
#extract all projects
# xpath cheatsheet: https://devhints.io/xpath
projects = browser.find_elements("xpath","//h1[@class='h3 lh-condensed']")
project_list = {}
for proj in projects:
    proj_name = proj.text #name
    proj_url = proj.find_elements("xpath","a")[0].get_attribute('href') #project URL
    project_list[proj_name] = proj_url
    
#quit once done scraping
browser.quit()
```

```python
df_project = pd.DataFrame.from_dict(project_list,orient='index')
df_project['project_name'] = df_project.index
df_project.columns = ['project_url','project_name']
df_project = df_project.reset_index(drop= True)
df_project
```

```python
#export to CSV for future use
df_project.to_csv('project_list.csv')
```
